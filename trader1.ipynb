```python
import yfinance as yf    
#shelanskey
import pandas as pd

dataF = yf.download("EURUSD=X", start="2024-06-06", end="2024-06-16", interval='15m') #YYYY=MM=DD
dataF.iloc[:,:]
#dataF.Open.iloc
```

```python
def signal_generator(df):
    open = df.Open.iloc[-1]
    close = df.Close.iloc[-1]
    previous_open = df.Open.iloc[-2]
    previous_close = df.Close.iloc[-2]

    #bearish
    if(open>close and previous_open<previous_close and close<previous_open and open>=previous_close):
        return 1

    #bullish
    elif(open<close and previous_open>previous_close and close>previous_open and open<=previous_close):
        return 2
    
    else: #no pattern
        return 0
    
signal = []
signal.append(0)
for i in range(1,len(dataF)):
    df = dataF[i-1:i+1]
    signal.append(signal_generator(df))

dataF["signal"] = signal
```
